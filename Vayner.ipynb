Question 2: What is the Cost per video view and Cost per completion for the Free Shipping + Free Shipping CTA video creatives (please filter by video post type)

```python
filtered = facebook_df[
    (facebook_df["Creative"] == "Free Shipping + Free Shipping CTA") &
    (facebook_df["Post Type"] == "Video")]

total_spend = filtered["Media Spend"].sum()
total_video_views = filtered["Video Views"].sum()
total_completions = filtered["Video Views to 95%"].sum() 

cpv = total_spend / total_video_views
cpcmp = total_spend / total_completions

cpv, cpcmp
```

Question 4: Please join the Site data to the Facebook data. Which audience type generated a ROAS of 5.46?

Question 5: If CTR increased 150% on a given audience type but CPC remained the same, what happened to its CPM?

Question 6: Our campaign recently doubled its attributed revenue on half of its typical purchase volume, while media spend remained the same in the same time period. What happened to ROAS and Average Order Value (AOV)?

Question 8: Of all days media was live, which day had the lowest ROAS?

```python
merged = facebook_df.merge(site_df[["Placement ID", "Total Revenue"]], on = "Placement ID", how = "left")

merged["ROAS"] = merged["Total Revenue"] / merged["Media Spend"]
merged["ROAS"] = merged["ROAS"].fillna(0)

daily_roas = merged.groupby("Date")["ROAS"].mean().sort_values()

lowest_day = daily_roas.idxmin()
lowest_roas = daily_roas.min()

print("Day with Lowest ROAS:", lowest_day)
print("Lowest ROAS Value:", lowest_roas)
```

Question 9: How much more media spend would be required to drive 70,000 total clicks (inclusive of clicks already generated) within the Valentine’s Day Gift Offer campaign, assuming a $13 CPM and a 1.20% CTR going forward?

```python
target_clicks = 70000
CTR = 0.012
CPM = 13
valentine_df = facebook_df[facebook_df["Campaign"] == "Valentine's Day Gift Offer"]

current_clicks = valentine_df["Clicks"].sum()
needed_clicks = target_clicks - current_clicks

needed_impressions = needed_clicks / CTR

additional_spend = (needed_impressions / 1000) * CPM

additional_spend
```
